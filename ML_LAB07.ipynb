```python
import pandas as pd
import numpy as np
from collections import Counter 
from collections import defaultdict

dic_pos = defaultdict(float)
dic_neg = defaultdict(float)
```

```python
str1 = ['I love mangoes','Mangoes are a lot tastier than other fruits','I like mangoes because they are good',
        'Apples are good too but I prefer mangoes','People say mango is the king of fruit so that is why mangoes are good',
        'Mangoes are good because they are juicy']

str2 = ['Eating too many mangoes can have a negative impact on your health','Mangoes are good but not too good I like apples more than mangoes',
        'I don’t like mangoes because they are expensive','Mangoes are not good','I don’t like mangoes']

len_str1=0
len_str2=0
for i in range(len(str1)):
    len_str1 +=len(str1[i].split())

for i in range(len(str2)):
    len_str2 +=len(str2[i].split())


######################################  
    
unique = []   
unique1 = []
for i in range(len(str1)):
    x = str1[i].lower()
    x = x.split()
    for j in x:
        if j not in unique1:
            unique1.append(j)
        if j not in unique:
            unique.append(j)
            
unique2 = []
for i in range(len(str2)):
    x = str2[i].lower()
    x = x.split()
    for j in x:
        if j not in unique2:
            unique2.append(j)
        if j not in unique:
            unique.append(j)

print("|v| = ",len(unique))
print("|c1| = ",len_str1)
print("|c2| = ",len_str2)
print()
print("unique words for both strings\n",unique) 

print("\nunique words for string 1")
print(unique1)

print("\nunique words for string 2")
print(unique2)

########   probabilities for pos class   ########

print('\n\n***************************************************\n\n')

def prob_of_positive_class():

    p=0
    pos_class=[]
    for i in unique1:
        for j in range(len(str1)):
            t = str1[j].lower()
            t = t.split()
            for k in t:
                if i == k:
                    p+=1
    #     dic[i]=p
        pos_class.append(p)
        p=0  

    prob_pos=[]
    for i in pos_class:
        pr = (i+1)/(len_str1+len(unique))
        prob_pos.append(pr)
    counter=0
    for i in unique1:
        dic_pos[i]=prob_pos[counter]
        counter+=1
    print("\nfinal prob of pos\n",prob_pos)
    print('\n',dic_pos)
    print('\n\n***************************************************\n\n')
    
def prob_of_negative_class():

    p=0
    neg_class=[]
    for i in unique2:
        for j in range(len(str2)):
            t = str2[j].lower()
            t = t.split()
            for k in t:
                if i == k:
                    p+=1
    #     dic[i]=p
        neg_class.append(p)
        p=0  

    prob_neg=[]
    for i in neg_class:
        pr = (i+1)/(len_str2+len(unique))
        prob_neg.append(pr)
    counter=0
    for i in unique2:
        dic_neg[i]=prob_neg[counter]
        counter+=1
    print("\nfinal prob of neg\n",prob_neg)
    print('\n',dic_neg)
    print('\n\n***************************************************\n\n')

prob_of_positive_class()
prob_of_negative_class()

exception_pos = 1/(len_str1+len(unique))
exception_neg = 1/(len_str2+len(unique))
```

```python
def find(user):
    
    prob_pos=[]
    prob_neg=[]
    for k in user:

        x = dic_pos[k]
        if x!=0:   
            prob_pos.append(x)
        else:
            prob_pos.append(exception_pos)

        x = dic_neg[k]
        if x!=0:   
            prob_neg.append(x)
        else:
            prob_neg.append(exception_neg)

    prob_pos.append(6/11)
    prob_neg.append(5/11)
    x1 = np.prod(prob_pos)
    x2 = np.prod(prob_neg)

    if x1>x2:
        print('positive')
    else:
        print('negative')
        
```
