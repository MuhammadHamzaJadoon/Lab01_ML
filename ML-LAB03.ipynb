# Projectile of Orthogonal Vector

```python
mag = np.sqrt(sum(x1**2))
proj = ((np.dot(x2,x1)*x1/mag))
```

```python
x2 = np.array(x2)
x2_m = np.sqrt(sum(x2**2))
proj_2 = ((np.dot(x2,x1)*x2/x2_m))
```

```python
fig=plt.figure(figsize=(10,10))
gr=fig.add_subplot(111)
gr.grid(alpha=0.4)
gr.set(xlim=(-8,8), ylim=(-8,8))
gr.annotate('', xy=x1, xytext=(0,0), arrowprops=dict(facecolor='b'))
gr.annotate('', xy=x2, xytext=(0,0), arrowprops=dict(facecolor='g'))
gr.annotate('', xy=proj, xytext=(0,0), arrowprops=dict(facecolor='r'))
gr.annotate('', xy=proj_2, xytext=(0,0), arrowprops=dict(facecolor='y'))
plt.show()
```

# Projection

```python
x1_mag = np.sqrt(np.sum(x1*x1))
px1 = np.dot(x1 , x2)*x1/x1_mag
px1
```

```python
x2 = np.array(x2)
x2_mag = np.sqrt(np.sum(x2*x2))
px2 = np.dot(x2 , x1)*x2/x2_mag
px2
```

```python
fig=plt.figure(figsize=(10,10))
gr=fig.add_subplot(111)
gr.grid(alpha=0.4)
gr.set(xlim=(-8,8), ylim=(-8,8))
gr.annotate('', xy=x1, xytext=(0,0), arrowprops=dict(facecolor='b'))
gr.annotate('', xy=x2, xytext=(0,0), arrowprops=dict(facecolor='g'))
gr.annotate('', xy=px1, xytext=(0,0), arrowprops=dict(facecolor='r'))
gr.annotate('', xy=px2, xytext=(0,0), arrowprops=dict(facecolor='y'))
plt.show()
```

```python
x3 = np.random.randint(1,6,(5,2))[0]
x4 = np.random.randint(1,6,(5,2))[0]
x3_mag = np.sqrt(np.sum(x3*x3))
x4_mag = np.sqrt(np.sum(x4*x4))
px3 = np.dot(x3 , x4)*x3/x3_mag
px4 = np.dot(x3 , x4)*x4/x4_mag
```
