# Task-2 (ii)

```python
names = ["hamza", "ali", "qasim","xass","qwqw","xqqq"]
print("unsorted : ",names)
names.sort()
print("sorted : ",names)
c=0
delete =[]
a_names=''
for i,j in enumerate(names):

    if j[0]=='x':
        delete.append(i)
        
# print(delete)
for i in delete:
    a_names+=names[i]
    
for i in reversed(delete):
    names.pop(i)
    
names.append(a_names)
print("final : ",names)

```
