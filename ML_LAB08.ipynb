```python
import pandas as pd
import numpy as np
from itertools import chain
import math

df = pd.read_csv("play_tennis.csv")
df = df.drop("day",axis=1)
arr = df.values.tolist()
```

```python
# -entropy_label

cols = df.columns.tolist()
feature_vals=[]
for i in cols:
    feature_vals.append(df[i].unique().tolist())
    
play = feature_vals[-1:]
play = list(chain.from_iterable(play))
feature_vals.pop(-1)
```

```python
# IG(wind) = entropy of label - entropy(weak)*prob(weak) + entropy(strong)*prob(strong)
# IG(humidity) = entropy of label - entropy(high)*prob(high) + entropy(normal)*prob(normal) 

no=0
yes=0
length = len(df['play'])
for i in range(len(df['play'])):
    
    if play[0] == df['play'][i]:
        no+=1
    if play[1] == df['play'][i]:
        yes+=1
        
entropy_label = no/length*math.log2(no/length) + yes/length*math.log2(yes/length)
```

```python
def entropy():
    
    p1=0
    p2=0
    total=0
    ent = []
    for k in range(len(feature_vals)):

        arr = []
        for j in range(len(feature_vals[k])):

            l=[]
            for i in range(len(df['play'])):

                if df[cols[k]][i] == feature_vals[k][j] and df['play'][i] == "Yes":
                    p1+=1
                if df[cols[k]][i] == feature_vals[k][j] and df['play'][i] == "No":
                    p2+=1  

                if df[cols[k]][i] == feature_vals[k][j]:
                    total+=1

            l.append(p1)
            l.append(p2)
            l.append(total)
            arr.append(l)
            p1=0
            p2=0
            total=0

        ent.append(arr)
        
    return ent

ent = entropy()
print(ent)
```

```python
entropy = []
probability = []

for i in range(len(feature_vals)):
    
    en = []
    pr = []
    add=0
    for j in range(len(feature_vals[i])):

        add += ent[i][j][2]
        try:
            en.append(ent[i][j][0]/ent[i][j][2]*(-math.log2(ent[i][j][0]/ent[i][j][2])) + ent[i][j][1]/ent[i][j][2]*(-math.log2(ent[i][j][1]/ent[i][j][2])))
        except:  
            try:
                en.append(ent[i][j][0]/ent[i][j][2]*(-math.log2(ent[i][j][0]/ent[i][j][2])))
            except:
                en.append(ent[i][j][1]/ent[i][j][2]*(-math.log2(ent[i][j][1]/ent[i][j][2])))
    
    for j in range(len(feature_vals[i])):
        pr.append(ent[i][j][2]/add)
         
    entropy.append(en)
    probability.append(pr)
```

```python
# entropy_label
r=0
vals=[]
for i in entropy:
    x=[]
    c=0
    for j in i:
        x.append(j * probability[r][c])
        c+=1
    vals.append(x)
    r+=1
    
final=[]
for i in vals:
    value=0
    for j in i:
        if value == 0:
            value=j
        else:
            value = value + j
    
    final.append(-entropy_label - value)
    
# print('info gain of all four columns : ',final)
print('Maximum info gain is',final[0],'for outlook')
```
