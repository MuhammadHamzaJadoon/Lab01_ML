```python
index=[0]
training_1_feature=[]
training_2_feature=[]
training_3_feature=[]
training_4_feature=[]

for i in range(24):
    x = np.delete(training_1[i],index)
    training_1_feature.append(x)
    x = np.delete(training_2[i],index)
    training_2_feature.append(x)
    x = np.delete(training_3[i],index)
    training_3_feature.append(x)
    x = np.delete(training_4[i],index)
    training_4_feature.append(x)
```

```python
label_1=[]
label_2=[]
label_3=[]
label_4=[]

x_train_1=[]
y_train_1=[]
x_train_2=[]
y_train_2=[]
x_train_3=[]
y_train_3=[]
x_train_4=[]
y_train_4=[]

for i in range(24):
    
    label_1.append(training_1[i][0])
    label_2.append(training_2[i][0])
    label_3.append(training_3[i][0])
    label_4.append(training_4[i][0])
    
training_1_label=np.array(label_1)
training_2_label=np.array(label_2)
training_3_label=np.array(label_3)
training_4_label=np.array(label_4)

x_train_1=training_1_feature+training_2_feature+training_3_feature
y_train_1=np.concatenate((training_1_label,training_2_label,training_3_label))

x_train_2=training_1_feature+training_2_feature+training_4_feature
y_train_2=np.concatenate((training_1_label,training_2_label,training_4_label))

x_train_3=training_1_feature+training_3_feature+training_4_feature
y_train_3=np.concatenate((training_1_label,training_3_label,training_4_label))

x_train_4=training_2_feature+training_3_feature+training_4_feature
y_train_4=np.concatenate((training_2_label,training_3_label,training_4_label))

```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import  accuracy_score
```

```python
manual_accuracy=[]

knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_1,y_train_1)
predictions=knn.predict(training_4_feature)
# print(predictions)
knn.score(training_4_feature,training_4_label)
manual_accuracy.append(accuracy_score(training_4_label,predictions))
##############################################################

knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_2,y_train_2)
predictions=knn.predict(training_3_feature)
# print(predictions)
knn.score(training_3_feature,training_3_label)
manual_accuracy.append(accuracy_score(training_3_label,predictions))
##############################################################

knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_3,y_train_3)
predictions=knn.predict(training_2_feature)
# print(predictions)
knn.score(training_2_feature,training_2_label)
manual_accuracy.append(accuracy_score(training_2_label,predictions))
##############################################################

knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_4,y_train_4)
predictions=knn.predict(training_1_feature)
# print(predictions)
knn.score(training_1_feature,training_1_label)
manual_accuracy.append(accuracy_score(training_1_label,predictions))
##############################################################
```

```python
#################################################################################
X=training_1_feature+training_2_feature+training_3_feature+training_4_feature
y=np.concatenate((training_1_label,training_2_label,training_3_label,training_4_label))
```

```python
from sklearn.model_selection import cross_val_score,KFold
from sklearn.model_selection import train_test_split

# X_train, X_test, y_train, y_test = train_test_split(X, y ,random_state=42,stratify=y)
kf=KFold(n_splits=4,shuffle=True,random_state=42)
knn=KNeighborsClassifier(n_neighbors=5)
cv_results=cross_val_score(knn,X,y,cv=kf)
```
